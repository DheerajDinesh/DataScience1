```python
x_train=df.iloc[:,:-1].values
x_train
```

```python
y_train=df.iloc[:,-1].values
y_train
```

```python
df1=pd.read_csv('/content/drive/MyDrive/Disease_Prediction/Testing.csv')
df1
```

```python
x_test=df1.iloc[:,:-1].values
x_test
```

```python
y_test=df1.iloc[:,-1].values
y_test
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
from sklearn.model_selection import GridSearchCV
knn1=KNeighborsClassifier()
param_knn={'n_neighbors':[3,5,7,9],'weights':['uniform','distance'],'algorithm':['auto','ball_tree','kd_tree','brute']}
clf_knn=GridSearchCV(knn1,param_knn,cv=10,scoring='accuracy')
clf_knn.fit(x_train,y_train)
```

```python
print(clf_knn.best_params_)
```

```python
tree1=DecisionTreeClassifier()
param_tree={'criterion':['gini','entropy','log_loss'],'splitter':['best','random']}
clf_tree=GridSearchCV(tree1,param_tree,cv=10,scoring='accuracy')
clf_tree.fit(x_train,y_train)
```

```python
print(clf_tree.best_params_)
```

```python
knn=KNeighborsClassifier(n_neighbors=3,algorithm='auto',weights='uniform')
base=GaussianNB()
model=SVC()
dec_tree=DecisionTreeClassifier(criterion='entropy',splitter='best')
lst=[knn,base,model,dec_tree]
```

```python
for i in lst:
  print("The model is", i)
  print('*'*100)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print(y_pred)
  print('*'*100)
  print('Confusion Matrix is')
  print(confusion_matrix(y_test,y_pred))
  print('*'*100)
  print('Accuracy Score =', accuracy_score(y_test,y_pred))
  print('*'*100)
  print(classification_report(y_test,y_pred))
```
